```python
import pandas as pd
import numpy as np
```

```python
US_only = pd.read_csv("./Data/Project Movie Data (For Students)/title-akas-us-only.csv", low_memory=False)
```

```python
display(US_only)
```

```python
basics = pd.read_csv("./Data/Project Movie Data (For Students)/title.basics.tsv.gz", sep='\t', low_memory=False)
```

```python
display(basics)
```

```python
# Filter the basics table down to only include the US by using the filter akas dataframe
filter_us_titles = basics['tconst'].isin(US_only['titleId'])
basics = basics[filter_us_titles]
basics

```

```python
basics = basics.replace({'\\N':np.nan})

```

```python
display(basics)
```

```python
basics = basics.dropna(subset=['runtimeMinutes','genres'])
```

```python
display(basics)
```

```python
basics = basics[basics['titleType'] == 'movie']
```

```python
display(basics)
```

```python
basics['startYear'] = basics['startYear'].astype(float)
```

```python
basics = basics[(basics['startYear'] >= 2000) & (basics['startYear'] <= 2022)]
```

```python
display(basics)
```

```python
basics = basics[~basics['genres'].str.contains("Documentary")]
```

```python
display(basics)
```

```python
basics.info()
```

```python
basics.head()
```

```python
folder_path = './Data/Project Movie Data (For Students)/'
file_path = folder_path + 'Basics.csv'
basics.to_csv(file_path, index=False)
```

```python
ratings = pd.read_csv("./Data/Project Movie Data (For Students)/title.ratings.tsv.gz", sep='\t', low_memory=False)
```

```python
display(ratings)
```

```python
# Filter the basics table down to only include the US by using the filter akas dataframe
filter_us = ratings['tconst'].isin(basics['tconst'])
ratings = ratings[filter_us]
ratings
```
